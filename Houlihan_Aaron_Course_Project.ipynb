```python
### Part 2. Correlation ###
### Question 1. Pearson and Spearman Correlation Coefficients ###

# Need to redefine variable lists to ensure all lists are same length (for pearson and spearman correlation coefficient calculations)
import sys
import numpy as np

file = open(input()) # Open and read input file

# Set variables (true/false and counters)
first = True
cnt = 0
cnt2 = 0

# Create empty lists for each variable
# Independent variables
temp = []
pH = []
DO = []
Biovolume = []
# Dependent variable
fishnum = []

for line in file:
    # Header row
    if first == True:
        data = line.split("\t")
        cnt = cnt + 1 # Multiple "header" rows for this dataset (separate rows for units)
        if cnt >= 3: # Groups all of these non-data rows together into header category
            first = False
    # Data rows
    else:
        data = line.split("\t")
        if cnt2 == 72 or cnt2 == 75: # Skipping empty rows (no data was entered in these rows of the file)
            pass
        elif cnt2 == 111: # Readjusting fish num (these rows contained a range of values, median of each range was used here)
            data[72] = 27.5
            temp.append(float(data[27])) 
            pH.append(float(data[29]))
            DO.append(float(data[31]))
            Biovolume.append(float(data[65]))
            fishnum.append(float(data[72]))
        elif cnt2 == 115 or cnt2 == 119:
            data[72] = 17.5
            # Store each data value in its respective variable list
            temp.append(float(data[27])) 
            pH.append(float(data[29]))
            DO.append(float(data[31]))
            Biovolume.append(float(data[65]))
            fishnum.append(float(data[72]))
        else: # All other rows
            temp.append(float(data[27])) 
            pH.append(float(data[29]))
            DO.append(float(data[31]))
            Biovolume.append(float(data[65]))
            fishnum.append(float(data[72]))
    cnt2 = cnt2 + 1
    
from scipy import stats

Var = [temp, pH, DO, Biovolume, fishnum] # Store lists of variables within another list
cnt1 = 1
cnt2 = 0.85

# Pearson Correlation Coefficient Calculation
for a in Var:
    for b in Var:
        if a != b: # Do not compare variable against itself (such as temperature vs. temperature)
            PCC = stats.pearsonr(a, b) # Find pearson correlation coefficient between a and b
            # Counter system to prevent repeats
            if cnt2 > cnt1:
                cnt1 = cnt1 + 1
                pass
            else:
                print("Pearson Correlation Coefficient:", PCC.statistic, "Variable 1:", Var.index(a), "Variable 2:", Var.index(b)) # See key below to determine pair of variables associated with correlation coefficient
    cnt2 = cnt2*2

print("\n")
# Reset counter variables for spearman calculation
cnt1 = 1
cnt2 = 0.85

# Spearman Correlation Coefficient Calculation
for a in Var:
    for b in Var:
        if a != b: # Do not compare variable against itself (such as temperature vs. temperature)
            SCC = stats.spearmanr(a, b) # Find spearman correlation coefficient between a and b
            # Counter system to prevent repeats
            if cnt2 > cnt1:
                cnt1 = cnt1 + 1
                pass
            else:
                print("Spearman Correlation Coefficient:", SCC.statistic, "Variable 1:", Var.index(a), "Variable 2:", Var.index(b)) # See key below to determine pair of variables associated with correlation coefficient
    cnt2 = cnt2*2

print("\n")

### Question 2. Linear Regression ###

# Convert list of variables into an array (for use of linear regression function)
Vararray = np.array(Var)

# Reset counter variables for linear regression calculation
cnt1 = 1
cnt2 = 0.85
cnta = 0
cntb = 0

import matplotlib.pyplot as plt

# Double for loop to iterate through all possible variable combinations
for a in Vararray:
    for b in Vararray:
        if cnta != cntb: # Avoid repeats
            result = stats.linregress(a, b) # Linear regression functio
            # Counter system to prevent repeats
            if cnt2 > cnt1:
                cnt1 = cnt1 + 1
                pass
            else:
                print(result, "Variable 1:", cnta, "Variable 2:", cntb, "\n") # Cannot index array like you can a list, so use separate counter to keep track of which variables are being compared
                ### Question 3. Scatter Plots with Linear Regression Lines Overlaid ###
                plt.scatter(a, b) # Scatter plot
                m, c = np.polyfit(a, b, 1) # Fit regression line
                y = m*np.array(a) + c # Equation of the line
                plt.plot(a, y, color = "green") # Plot the line
                plt.xlabel(cnta) # x-axis label
                plt.ylabel(cntb) # y-axis label
                plt.title("Linear Regression of Select Variables") # Title
                plt.show() # Needed to show each plot separately
        cntb = cntb + 1
    cnt2 = cnt2*2
    cntb = 0
    cnta = cnta + 1
    
# Key to distinguish variables    
print("\nVariables Key:")
print("Temperature (C) = 0")
print("pH = 1")
print("Dissolved Oxygen (mg/L) = 2")
print("Biovolume (mL) = 3")
print("Fish Count = 4")
```

```python
### Part 3. Dimensionality Reduction ###
### Question 1. Principal Component Analysis ###

import sys
import numpy as np

file = open(input()) # Open and read input file

# Set variables (true/false and counters)
first = True
cnt = 0
cnt2 = 0

# Create empty list for eventual list of lists (lol)
# Instead of separating the data by column, want to separate it by row
lol = []
intermediate = []

for line in file:
    # Header row
    if first == True:
        data = line.split("\t")
        cnt = cnt + 1 # Multiple "header" rows for this dataset (separate rows for units)
        if cnt >= 3: # Groups all of these non-data rows together into header category
            first = False
    # Data rows
    else:
        data = line.split("\t")
        if cnt2 == 72 or cnt2 == 75: # Readjusting empty rows (no data was entered in these rows of the file)
            data[27] = 0
            data[29] = 0
            data[31] = 0
            data[65] = 0
        if cnt2 == 111: # Readjusting fish num (these rows contained a range of values, median of each range was used here)
            data[72] = 27.5
        elif cnt2 == 115 or cnt2 == 119:
            data[72] = 17.5
        # List of lists
        intermediate.append(data[27]) # Intermediate is list of 5 variables we want
        intermediate.append(data[29])
        intermediate.append(data[31])
        intermediate.append(data[65])
        intermediate.append(data[72])
        lol.append(intermediate) # lol is final storage place for each intermediate list
    cnt2 = cnt2 + 1
    intermediate = [] # Reset "Intermediate" each row

from sklearn.decomposition import PCA

# PCA Analysis
Vararray = np.array(lol) # Convert lol to array
Vararray = np.delete(Vararray, (69), axis = 0) # Remove outlier rows (rows in which there was no data so zeros were added)
Vararray = np.delete(Vararray, (71), axis = 0)
pca = PCA()
fit = pca.fit_transform(Vararray)

#print("Principal Component Analysis of each Row:\n",fit) # PCA of each row
print("Principal Component Weights:\n",pca.explained_variance_ratio_) # PCA weights

weights = pca.explained_variance_ratio_.tolist() # Convert PCA weights to list format to plot on bar graph

import matplotlib.pyplot as plt

### Question 2. Extract PCA Weights and Plot Using Bar Graph ###
plt.bar([1, 2, 3, 4, 5], weights)
plt.xlabel("Principal Component")
plt.ylabel("Weight (Decimal)")
plt.title("Principal Component Weights")
plt.show()

### Question 3. Visualize Data Using PCA1 and PCA2. Color Code by Relevant Categories.
pca2 = PCA(n_components = 2) # Only want PCA1 and PCA2
fit2 = pca2.fit_transform(Vararray)
print("[PC1, PC2]",pca2.explained_variance_ratio_)

# Color coding based on sampling area
plt.scatter(fit[:19, 0], fit[:19, 1], color = "yellow") # Yellow = ENP (Everglades National Park)
plt.scatter(fit[19:45, 0], fit[19:45, 1], color = "orange") # Orange = WCA3 (Water Conservation Area 3)
plt.scatter(fit[45:52, 0], fit[45:52, 1], color = "red") # Red = WCA2 (Water Conservation Area 2)
plt.scatter(fit[52:59, 0], fit[52:59, 1], color = "blue") # Blue = LNWR (Loxahatchee National Wildlife Refuge)
plt.scatter(fit[59:76, 0], fit[59:76, 1], color = "yellow")
plt.scatter(fit[76:104, 0], fit[76:104, 1], color = "orange")
plt.scatter(fit[104:109, 0], fit[104:109, 1], color = "red")
plt.scatter(fit[109:, 0], fit[109:, 1], color = "blue")
#plt.scatter(fit[:,0], fit[:,1]) # PCA1 vs. PCA2
plt.xlabel("PC1 (" + str(np.round(pca2.explained_variance_ratio_[0]*100, decimals=2)) + "%)") # Principal component with corresponding weight
plt.ylabel("PC2 (" + str(np.round(pca2.explained_variance_ratio_[1]*100, decimals=2)) + "%)")
plt.title("PC1 vs. PC2")
plt.show()

print("\nKey:")
print("Yellow = Everglades National Park")
print("Orange = Water Conservation Area 3")
print("Red = Water Conservation Area 2")
print("Blue = Loxahatchee National Wildlife Refuge")
```
